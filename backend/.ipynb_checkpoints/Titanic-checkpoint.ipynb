<h2>Filling Missing Values</h2>

```python
train_df.Age.mean()
```

```python
np.median(train_df.Age)
```

```python
age_median = train_df['Age'].median()
fare_mean = train_df['Fare'].mean()
embarked_mode = train_df['Embarked'].mode()[0]
```

```python
train_df = train_df.assign(Age=train_df['Age'].fillna(age_median))
test_df = test_df.assign(Age=test_df['Age'].fillna(age_median))  
```

```python
test_df = test_df.assign(Fare=test_df['Fare'].fillna(fare_mean))
train_df = train_df.assign(Embarked = train_df['Embarked'].fillna(embarked_mode))
```

```python
print(f"Sum of missing value for train set {train_df.isnull().any().sum()}")
print(f"Sum of missing value for test set {test_df.isnull().any().sum()}")
```

```python
# Log transform Fare (common fix for right-skewed data)
train_df = train_df.assign(Fare=np.log1p(train_df['Fare']))
test_df = test_df.assign(Fare=np.log1p(test_df['Fare']))
```

```python
X = train_df.drop(columns=['Survived'])
y = train_df.Survived
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('ordinal_sex', OrdinalEncoder(), ['Sex']),                                        
    ('ohe_embarked', OneHotEncoder(drop='first', sparse_output=False), ['Embarked'])   
], remainder='passthrough')  # keeps all other columns

# Apply
X_train = preprocessor.fit_transform(X)
X_test = preprocessor.transform(test_df)  # never fit_transform on test!
```

```python
#Spliting data into train and val set

X_train, X_val, y_train, y_val = train_test_split(X_train, y, test_size=.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # fit only on train
X_val_scaled = scaler.transform(X_val)          # transform only
X_test_scaled = scaler.transform(X_test)        # same for test
```

<h2>Model Building</h2>

```python
models = {
    "RF": RandomForestClassifier(),
    "LM": LogisticRegression()
}
result = {}
items = []

for val, model in models.items():
    mod = model.fit(X_train_scaled, y_train)
    result[f'{val}_score'] = mod.score(X_val_scaled, y_val)
    items.append(result)

print(result)
```

```python
models['RF'].predict(X_val)
```

```python
y_pred = models['RF'].predict(X_test_scaled)
```

```python
# ✅ Correct
y_val_pred = model.predict(X_val_scaled)
print(classification_report(y_val, y_val_pred))
```

```python
sb.heatmap(confusion_matrix(y_val, y_val_pred), annot=True)
```
